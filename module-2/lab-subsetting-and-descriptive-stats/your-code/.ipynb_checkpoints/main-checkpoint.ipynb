# Challenge 3

#### Open the Orders.csv dataset. Name your dataset orders

```python
orders = pd.read_csv('Orders.csv')
orders.head()
```

#### Explore your dataset by looking at the data types and the summary statistics. Comment your results

```python
'''Todos aquellos valores que no son numércios, son type "object"'''
orders.dtypes
```

```python
#comentario
round(orders.describe())
```

####  What is the average Purchase Price?

```python
orders['amount_spent'].mean()
```

#### What were the highest and lowest purchase prices? 

```python
orders['amount_spent'].max()
```

```python
orders['amount_spent'].min()
```

#### Select all the customers we have in Spain

```python
spain_customers = orders[orders['Country'] == 'Spain']['CustomerID'].unique()
spain_customers
```

#### How many customers do we have in Spain?
##### Hint : Use value_counts() (NO funciona)

```python
len(spain_customers)
```

#### Select all the customers who have bought more than 50 items ?

```python
#Hay que ponerle unique para no considerarlos más de una vez.
more_50 = orders[orders.Quantity > 50]['CustomerID'].unique()
len(more_50)
```

#### Select orders from Spain that are above 50 items

```python
orders[(orders.Country == 'Spain') & (orders.Quantity > 50)]
```

#### Select all free orders

```python
orders[orders.amount_spent == 0]
```

#### Select all orders that are 'lunch bag'
#### Hint: Use string functions

```python
orders.loc[orders['Description'].apply(lambda x: 'lunch bag' in x)]
```

#### Select all orders that are made in 2011 and are 'lunch bag'  

```python
orders[['Description','year']].loc[
    (orders.Description.apply(lambda x: 'lunch bag' in x))
    & 
    (orders.year==2011)]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
orders[orders['Country']=='Spain'].hist('amount_spent');
```

#### Select all orders made in the month of August

```python
orders[orders['month']==8]
```

#### Select how many orders are made by countries in the month of August
##### Hint: Use value_counts()

```python
orders[['Quantity','Country']].loc[
    orders['month']==8
    ].groupby('Country').count()
```

#### What's the  average amount of money spent by country

```python
orders[orders['month']== 8].groupby('Country')[
                     ['Country', 'amount_spent']
                                              ].mean()
```

#### What's the most expensive item?

```python
orders[orders['UnitPrice'] == orders['UnitPrice'].max()]
```

#### What was the average amount spent per year ?

agg aplica una función a los elementos definidos. Por eso el resultado sólo es ese promedio y no toda la fila.

```python
orders.groupby('year').agg({'amount_spent':np.mean})
```
