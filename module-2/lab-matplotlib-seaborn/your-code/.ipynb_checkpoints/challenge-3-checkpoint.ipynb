#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# plot the histogram
titanic['Age'].hist(bins=60);
# set labels
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Count');
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# bins
bins = [10,20,50]

# subplot creation
f, ax = plt.subplots(1,3)
f.set_figwidth(15)

# plot the histogram
for i in range(len(bins)):
    plt.sca(ax[i])
    titanic['Age'].hist(bins=bins[i])
    plt.title('n=' + str(bins[i]))
    plt.xlabel('Age')
    plt.ylabel('Count')
```

#### Whether the bin size affects your plot? Why?

```python
#your comment here
# It affects the precision. With more bins there is more precision
# for a more general approach less bins is better
```

#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way: MATPLOTLIB
titanic['Gender'].value_counts().plot(kind ='bar')
plt.xlabel('Gender')
plt.ylabel('People')
plt.xticks(rotation = 0)
plt.show()
```

```python
# your code here- 2nd way: SEABORN
sns.countplot(x='Gender',data=titanic)
plt.show()
```

#### Use the appropriate plot to visualize the column `Pclass` 


```python
# Barplot is the best option
titanic['Pclass'].value_counts().plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('People')
plt.xticks(rotation=0)
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way: SCATTERPLOT
plt.scatter(titanic['Fare'],titanic['Age'])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()
```

```python
# your code-2nd way: SEABORN (JOINTPLOT)
sns.jointplot(x = 'Fare',y = 'Age',data = titanic)
plt.show()
```
