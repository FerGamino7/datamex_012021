# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
# Preparing data
totals = data.groupby('ItemType')[['RetailSales','RetailTransfers','WarehouseSales']].sum()
```

```python
# Plotting
totals.plot(kind='bar',figsize =(15,9));
plt.title('Total: Retail Sales, Retail Transfers and Warehouse Sales')
plt.ylabel('Total');
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
# Data preparation
supplier_sales = data.groupby('Supplier')[['WarehouseSales','RetailTransfers','RetailSales']].sum()
supplier_sales['Total'] = supplier_sales.sum(axis = 1)
top10_suppliers = supplier_sales.sort_values(by = 'Total', ascending = False)[:10]
top10_suppliers.drop('Total', axis = 1, inplace = True)
```

```python
# Plotting
top10_suppliers.plot(kind = 'barh', figsize = (15,9))
plt.title('sales mix for the top 10 suppliers', fontweight = 600)
plt.xlabel('Total')
plt.ylabel('Supplier');
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
# Data Transformation
avg_sales = data.groupby(['Year','Month']).mean()
xticks = [f'{str(index[1])}\n{str(index[0])}' for index in data.groupby(['Year','Month']).groups.keys()]
```

```python
# Data Plotting
avg_sales.plot(figsize=(15,9))
plt.title('Average sales per month',fontweight = 800)
plt.xlabel('Date')
plt.ylabel('Average Sales')
plt.xticks(range(len(xticks)),xticks);
```

## 4. Plot the same information as above but as a bar chart.

```python
# barchart
avg_sales.plot(kind = 'bar',figsize=(15,9))
plt.title('Average sales per month',fontweight = 800)
plt.xlabel('Date')
plt.ylabel('Average Sales')
plt.xticks(range(len(xticks)),xticks);
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
# Data Transformation
retail_item = pd.pivot_table(data,
                             columns = 'ItemType',values = 'RetailSales',
                             index=['Year','Month'],aggfunc = sum)
xticks = [f'{str(index[1])}\n{str(index[0])}'for index in data.groupby(['Year','Month']).groups.keys()]
```

```python
# Plotting
retail_item.plot(figsize=(15,9))
plt.title('Total retail sale per item',fontweight = 700)
plt.xlabel('Date')
plt.ylabel('Total Retail Sales')
plt.xticks(range(len(xticks)),xticks,rotation=0);
```

## 6. Plot the same information as above but as a bar chart.

```python
# Barchart
retail_item.plot(figsize=(15,9), kind = 'bar')
plt.title('Total retail sale per item',fontweight = 700)
plt.xlabel('Date')
plt.ylabel('Total Retail Sales')
plt.xticks(range(len(xticks)),xticks,rotation=0);
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
#implot
scatter = sns.lmplot('RetailSales','RetailTransfers',data,hue = 'ItemType',fit_reg = False)
scatter.fig.set_figwidth(12)
scatter.fig.set_figheight(8)
plt.title('Retail Sales vs Retail Transfers',fontweight = 800);
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
#pairplot
sns.pairplot(data, hue='ItemType');
```
