# Sustituyendo valores

```python
attacks7['Sex'] = attacks7['Sex'].str.replace('M ','M')
attacks7['Sex'] = attacks7['Sex'].str.replace('.','F')
attacks7['Sex'] = attacks7['Sex'].str.replace('N','F')
attacks7['Sex'] = attacks7['Sex'].str.replace('lli','F')
print(set(attacks7['Sex']))
```

```python
attacks7['Fatal (Y/N)'] = attacks7['Fatal (Y/N)'].str.replace('y','Y')
attacks7['Fatal (Y/N)'] = attacks7['Fatal (Y/N)'].str.replace('2017','UNKNOWN')
attacks7['Fatal (Y/N)'] = attacks7['Fatal (Y/N)'].str.replace('M','N')
attacks7['Fatal (Y/N)'] = attacks7['Fatal (Y/N)'].str.replace('N ','N')
attacks7['Fatal (Y/N)'] = attacks7['Fatal (Y/N)'].str.replace(' N','N')

print(set(attacks7['Fatal (Y/N)']))
```

```python
attacks7['Type'] = attacks7['Type'].str.replace('Boatomg','Boating')
attacks7['Type'] = attacks7['Type'].str.replace('Questionable','Invalid')


print(set(attacks7['Type']))
```

```python
attacks7[['Sex','Fatal (Y/N)']] = attacks7[['Sex','Fatal (Y/N)']].fillna('UNKNOWN')
```

```python
attacks7['Date'] = attacks7['Date'].str.replace('Reported ','')
attacks7['Date'] = attacks7['Date'].str.replace('Summer of ','')
attacks7['Date'] = attacks7['Date'].str.replace('Summer ','')
attacks7['Date'] = attacks7['Date'].str.replace('Before ','')
attacks7['Date'] = attacks7['Date'].str.replace('Early ','')
attacks7['Date'] = attacks7['Date'].str.replace('Late ','')
attacks7['Date'] = attacks7['Date'].str.replace('Ca. ','')
attacks7['Date'] = attacks7['Date'].str.replace('Circa ','')
attacks7['Date'] = attacks7['Date'].str.replace('     ','') #6 espacios
attacks7['Date'] = attacks7['Date'].str.replace('  ','') # dos espacios


print(set(attacks7['Date']))
```
